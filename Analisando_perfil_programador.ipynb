```python

dados = dados[['Age', 'Gender', 'JobPref', 'JobRoleInterest', 'EmploymentField', 'JobRelocateYesNo', 'HoursLearning', 'MoneyForLearning']]
dados
```

```python

dados = dados.drop_duplicates()
```

```python

dados.describe(include = 'O')
```

```python
# 'Aprendizado ao Longo do Tempo'
dados.groupby('Age')['HoursLearning'].mean().plot(kind='line')
plt.title('apprenticeship in Time')
plt.xlabel('Age')
plt.ylabel('Media')
plt.show()
```

```python
# Gênero
dados['Gender'].value_counts().plot(kind='bar')
plt.title('Distribuição de Gênero')
plt.xlabel('Gênero')
plt.ylabel('Contagem')
plt.show()
```

```python
  # Trabalho
dados['JobPref'].value_counts().plot(kind='bar')
plt.title('Preferência de Trabalho')
plt.xlabel('Preferência de Trabalho')
plt.ylabel('Contagem')
plt.show()
```
